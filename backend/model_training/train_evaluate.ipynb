# Imports

```python
from roboflow import Roboflow
from dotenv import load_dotenv
import os
from ultralytics import YOLO
```

# Constants

```python
load_dotenv()
api_key = os.getenv("ROBOFLOW_API")
```

```python
classes = {
    0: 'potato',
    1: 'tomato',
    2: 'beef',
    3: 'pork',
    4: 'chicken',
    5: 'zucchini',
    6: 'yellow_onion',
    7: 'radish',
    8: 'eggplant',
    9: 'green_onion',
    10: 'red_cabbage',
    11: 'carrot',
    12: 'egg'
}
```

# Load dataset

```python
rf = Roboflow(api_key=api_key)
project = rf.workspace("seongmin").project("food-ingredients-pnisb")
version = project.version(1)
dataset = version.download("yolov8", location="datasets/food-ingredients-pnisb")
```

# Train model

```python
# Define and load the model
model = YOLO("yolov8n.pt")
dataset_config = 'data.yaml'

# Train the model
results = model.train(data=dataset_config, epochs=40, imgsz=416, device=[0], batch=16)
```

# Evaluate model

```python
model = YOLO("runs/detect/train/weights/best.pt")
```
